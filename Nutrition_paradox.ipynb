```python
import requests
import pandas as pd


#urls

url_ob_ad = "https://ghoapi.azureedge.net/api/NCD_BMI_30C"
url_ob_ch = "https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C"
url_mal_ad = "https://ghoapi.azureedge.net/api/NCD_BMI_18C"
url_mal_ch = "https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C"
```

```python
# obesity_adult

res_1 = requests.get(url_ob_ad)
df_ob_ad = res_1.json()
df_ob_ad.keys()
```

```python
df_ob_ad =pd.DataFrame(df_ob_ad['value'])
df_ob_ad['age_group'] = 'Adult'
```

```python
#obesity_child

res_2 = requests.get(url_ob_ch)
df_ob_ch = res_2.json()
df_ob_ch = pd.DataFrame(df_ob_ch['value'])
df_ob_ch['age_group'] = 'Child'
```

```python
df_ob_ch.head()
```

```python
#malnutrition_adult

res_3 = requests.get(url_mal_ad)
df_mal_ad = res_3.json()
df_mal_ad = pd.DataFrame(df_mal_ad['value'])
df_mal_ad['age_group'] = 'Adult'
df_mal_ad.head()
```

```python
#malnutrition_child


res_4 = requests.get(url_mal_ch)
df_mal_ch = res_4.json()
df_mal_ch = pd.DataFrame(df_mal_ch['value'])
df_mal_ch['age_group'] = 'Child'
df_mal_ch.head()
```

```python
def clean_and_standardize(df):
    # Retain only necessary columns
    df = df[['ParentLocation', 'Dim1', 'TimeDim', 'Low', 'High', 'NumericValue', 'SpatialDim', 'age_group']]
    
    # Rename columns
    df.columns = ['Region', 'Gender', 'Year', 'LowerBound', 'UpperBound', 'Mean_Estimate', 'Country', 'Age_Group']
    
    
    
    
    return df
```

```python
df_ob_ad = clean_and_standardize(df_ob_ad)
df_ob_ch = clean_and_standardize(df_ob_ch)
df_mal_ad = clean_and_standardize(df_mal_ad)
df_mal_ch = clean_and_standardize(df_mal_ch)
```

```python
# Create obesity table

cur.execute("""
    CREATE TABLE obesity (
        Year INT,
        Gender VARCHAR(10),
        Mean_Estimate FLOAT,
        LowerBound FLOAT,
        UpperBound FLOAT,
        Age_Group VARCHAR(20),
        Country VARCHAR(100),
        Region VARCHAR(100),
        CI_Width FLOAT,
        Obesity_Level VARCHAR(20)
    );
""")
```

```python
myconnection.commit()
```

```python
# Create malnutrition table

cur.execute("""
    CREATE TABLE  malnutrition (
        Year INT,
        Gender VARCHAR(10),
        Mean_Estimate FLOAT,
        LowerBound FLOAT,
        UpperBound FLOAT,
        Age_Group VARCHAR(20),
        Country VARCHAR(100),
        Region VARCHAR(100),
        CI_Width FLOAT,
        Malnutrition_Level VARCHAR(20)
    );
""")

myconnection.commit()
```

```python
#insert into obesity table

for index, row in df_obesity.iterrows():
    cur.execute("""
        INSERT INTO obesity (Year, Gender, Mean_Estimate, LowerBound, UpperBound,
            Age_Group, Country, Region, CI_Width, Obesity_Level)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Year'], row['Gender'], row['Mean_Estimate'], row['LowerBound'],
        row['UpperBound'], row['Age_Group'], row['Country'], row['Region'],
        row['CI_Width'], row['obesity_level']
    ))

myconnection.commit()
```

```python
#insert into malnutriton table

for index, row in df_malnutrition.iterrows():
    cur.execute("""
        INSERT INTO malnutrition (Year, Gender, Mean_Estimate, LowerBound, UpperBound,
            Age_Group, Country, Region, CI_Width, Malnutrition_Level)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Year'], row['Gender'], row['Mean_Estimate'], row['LowerBound'],
        row['UpperBound'], row['Age_Group'], row['Country'], row['Region'],
        row['CI_Width'], row['malnutrition_level']
    ))

myconnection.commit()
```

Malnutrition:

Slight overall decline over time, but with regional inequality
